```python
discrete_feature=[feature for feature in df.columns if len(df[feature].unique())<20 and feature]
print('Discrete Variables Count: {}'.format(len(discrete_feature)))
```

```python
continuous_feature=[feature for feature in df.columns if df[feature].dtype!='O' and feature not in discrete_feature]
print('Continuous Feature Count {}'.format(len(continuous_feature)))
```

```python
import scipy.stats as stat
import pylab
def plot_data(df,feature):
    plt.figure(figsize=(12,6))
    plt.subplot(1,2,1)
    df[feature].hist()
    plt.subplot(1,2,2)
    stat.probplot(df[feature],dist='norm',plot=pylab)
```

```python
plot_data(df,'Price')
plt.title('Price')
plot_data(df,'Age_08_04')
plt.title('Age_08_04')
plot_data(df,'KM')
plt.title('KM')
plot_data(df,'Weight')
plt.title('Weight')
```

```python
df1=df.copy()
df1[continuous_feature]=np.log(df1[continuous_feature])

plot_data(df1,'Price')
plt.title('Price')
plot_data(df1,'Age_08_04')
plt.title('Age_08_04')
plot_data(df1,'KM')
plt.title('KM')
plot_data(df1,'Weight')
plt.title('Weight')
```

# Model building

```python
import statsmodels.formula.api as smf
model = smf.ols('Price~Age_08_04+KM+Weight+HP+cc+Gears+Quarterly_Tax+Doors', data = df_new).fit()
```

```python
model.summary()
```

```python
from statsmodels.graphics.regressionplots import influence_plot
```

```python
influence_points=model.get_influence()
c, p_value=influence_points.cooks_distance
```

```python
fig = plt.figure(figsize = (16,9))
x = [0,1330]
y = [0.5,0.5]
plt.plot(x, y,color='darkred', linewidth=2)
y1 = [0.02,0.02]
plt.plot(x , y1, color = 'red', linewidth = 2)
plt.stem(np.arange(len(df_new)), np.round(c, 3))
plt.xlabel('Row index')
plt.ylabel("Cook's Distance")
plt.show()
```

```python
model_c = smf.ols("Price~cc", data=df_new).fit()
model_c.summary()
```

```python
model_cd = smf.ols("Price~cc+Doors", data=df_new).fit()
model_cd.summary()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
x = df_new[['Age_08_04','KM','HP','cc','Doors','Quarterly_Tax','Weight']]
VIF_cal = pd.DataFrame()
VIF_cal['features'] = x.columns
VIF_cal['VIF'] = [variance_inflation_factor(x.values,i) for i in range(len(x.columns))]
```
